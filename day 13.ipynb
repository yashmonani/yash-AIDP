```python

import matplotlib.pyplot as plt
import numpy as np
import skimage as ski
import cv2
import pandas as pd
import face_recognition as fr
```

```python
#face detection registeration
fname = 'features.csv'
try:
    df=pd.read_csv(fname)
except: 
    df = pd.DataFrame({'name':[], 'enc':[]})
df=pd.DataFrame({'name':[],'enc':[]})
df
fd = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
vid = cv2.VideoCapture(0)
counter = 0
names= []
feats= []
name = input("Enter your name: ")
while True:
    ack, img=vid.read()#acquire video camera
    if ack:
        faces = fd.detectMultiScale(img,1.2,5, minSize=(150,150))#image ko scale ,minimum features
        if len(faces) == 1:
            x,y,w,h = faces[0]
            face_img = img[y:y+h,x:x+w,:].copy()
            face_enc = fr.face_encodings(face_img)     
            if len(face_enc) == 1:
                counter +=1
                names += [name]
                feats += [face_enc[0].tolist()]
                f = pd.DataFrame({'name':[name], 'enc':face_enc})
                df = pd.concat([df,f],axis=0,ignore_index=True)
            if counter == 20:
                f=pd.DataFrame({'name':names,'enc':feats})
                df=pd.concat([df,f],axis=0,ignore_index=True)
                # df.to_csv('face_features.csv')
                df.to_csv(fname)
                print(counter)
                break    
        cv2.imshow('Preview',img)
        key = cv2.waitKey(1)
        if key == ord('x'): 
            break
cv2.destroyAllWindows()
vid.release()
```
